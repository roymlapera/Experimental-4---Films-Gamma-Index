```python
import wx
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def get_path(wildcard):
    app = wx.App(None)
    style = wx.FD_OPEN | wx.FD_FILE_MUST_EXIST
    dialog = wx.FileDialog(None, 'Open', wildcard=wildcard, style=style)
    if dialog.ShowModal() == wx.ID_OK:
        path = dialog.GetPath()
    else:
        path = None
    dialog.Destroy()
    return path
```

```python
filename_film = get_path('*.tif')



filename_film = filename_film.encode('utf-8')
print filename_film


film_array = cv2.imread(filename_film,-1)[:,:,2]/1000.0#.astype(np.float64)

```

```python
%matplotlib inline

plt.figure(figsize = (8, 8))

plt.imshow(film_array, cmap = 'seismic', vmin=0.0, vmax = 2.0, interpolation = 'none')

plt.colorbar(fraction=0.04, pad=0.04) # estos numeritos tocalos hasta que la colorbar tenga el mismo alto que la figura

plt.axis('off')

plt.show()
```
